```python
import rsa
import base64
#Generate public key and private key
public_Key, private_Key = rsa.newkeys(2048)
#print generated public key
print(public_Key,"\n\n")
#print generated private key
print(private_Key)
```

```python
#sign a message
your_message = 'some text for sign with private key'
sign = rsa.sign(your_message.encode(),
                private_Key,"SHA-256")
#convert to base64
Base64_String = str(base64.b64encode(sign),'utf-8')
print(Base64_String)
```

```python
#verify signature
if (rsa.verify(your_message.encode(), sign, public_Key)):
    print('signature is valid.')
```
